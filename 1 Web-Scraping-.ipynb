# **Web Scraping**


## Objectives


In this lab I will perform the following:


* Extract information from a given web site 
* Write the scraped data into a csv file.


## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data I need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in a web broswer and study the contents of the web page before I start to scrape.


Import the required libraries


```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a webpage
import csv                     # for saving into CSV
```

Download the webpage at the url


```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
data  = requests.get(url).text 
```

Create a soup object


```python
soup = BeautifulSoup(data,"html.parser")  # create a soup object using the variable 'data'
```

Scrape the `Language name` and `annual average salary`.


```python
table = soup.find("table")
rows = table.find_all("tr")

languages = []
salaries = []
for row in rows[1:]:  
    cols = row.find_all("td")
    language = cols[0].text.strip()
    salary = cols[1].text.strip()
    languages.append(language)
    salaries.append(salary)
```

Save the scrapped data into a file named *popular-languages.csv*


```python
with open("popular-languages.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Language", "Average Annual Salary"])  # header row
    for lang, sal in zip(languages, salaries):
        writer.writerow([lang, sal])
```
